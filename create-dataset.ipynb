```python
import pandas as pd
import numpy as np
```

```python
#training set for bodies
training_body = pd.read_csv('/kaggle/input/nlp-project/train_bodies.csv')
print(training_body.head(10))
```

```python
#training set for stances
training_stance = pd.read_csv('/kaggle/input/nlp-project/train_stances.csv')
print(training_stance.head(10))
```

```python
# Returns Article Body when a specific body_id is given
def id2body(body_id,body_data):
    for i in range(len(body_data['articleBody'])):
        if(body_data['Body ID'][i] == body_id):
            return (body_data['articleBody'][i])
```

```python
print(len(training_stance['Body ID']))
print(len(training_body['Body ID']))
```

```python
# we remove newline characters and change 's to is and 't to not
#for example it's changed to it is and 
def preprocess(s):
    s = s.replace("\n","")
    s = s.replace("\'s"," is")
    s = s.replace("n\'t"," not")
    s = s.replace("\'d"," would")
    return s
```

```python
headline_wordcount = []
body_wordcount = []
headline = []
body = []
stance = []
for i in range(len(training_stance['Headline'])):
    headline_clean = preprocess(training_stance['Headline'][i])
    body_clean = preprocess(id2body(training_stance['Body ID'][i],training_body))
    headline.append(headline_clean)
    body.append(body_clean)
    headline_wordcount.append(len(headline_clean.split()))
    body_wordcount.append(len(body_clean.split()))
    stance.append(training_stance['Stance'][i])
```

```python
print(len(headline)==len(body)==len(headline_wordcount)==len(body_wordcount)==len(stance))
print(len(headline))
```

```python
# Saving training set to csv file 
train_set = pd.DataFrame()
train_set['Headline'] = headline
train_set['Body'] = body
train_set['Stance'] = stance
train_set['Headline Word Count'] = headline_wordcount
train_set['Body Word Count'] = body_wordcount
train_set.to_csv("/kaggle/working/train_Set.csv")
```
